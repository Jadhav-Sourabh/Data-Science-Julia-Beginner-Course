```julia
using Pkg 
Pkg.add("Plots")
```

```julia
using Plots
plot()
```

```julia
x = 1:15;
y = rand(15);
```

```julia
plot(x, y)
```

```julia
z = rand(15);
# Combining the plot
plot!(x, z)
```

```julia
p = plot(x, y)
plot!(p, x, z)
```

```julia
# Adding the title
plot(x, y, title="One random line chart")
```

```julia
# Giving the label
plot(x, y, title="Random line chart", label = "Line chart 1")
```

```julia
# Axis labels
plot(x, y, xlabel = "X axis", ylabel = "Y axis", title = "Random Line Chart", label = "Line 1")
```

```julia
# Scatter Chart
scatter(x, y)
```

```julia
# Bar chart
bar(x, y)
```

```julia
# Histogram
histogram(y)
```

```julia
# Dashboard
p1 = plot(x, y)
p2 = scatter(x, y)
p3 = bar(x, y)
p4 = histogram(y)

plot(p1, p2, p3, p4, layout = (2, 2), legend = false)
```

```julia
y = rand(15, 4)
plot(x, y, layout = (4, 1))
```
